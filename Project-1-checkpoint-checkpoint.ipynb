```python
#Crime Grouped by Neighborhood
neighborhood_group=Crime_df.groupby("Analysis Neighborhood")
neighborhood=neighborhood_group["Incident Date"].count()
#print(neighborhood)
plt.figure(figsize=(20,10))
plt.bar(neighborhood.index, neighborhood.values)
plt.xticks(rotation=90)
plt.xlabel("Neighborhood")
plt.ylabel("Number of Crimes")
plt.title("Number of Crimes vs Neighborhood")
plt.grid()
plt.show()
```

```python
#Crime by Police Districts
districts=Crime_df["Police District"].value_counts()
#print(districts)

x_axis=np.arange(len(districts))
x_ticks=[x for x in x_axis]
#print(x_ticks)

districts_count=[value for value in districts.values]
#print(districts_count)

plt.bar(x_axis, districts_count, color="r", alpha=0.8)
plt.xticks(x_ticks, districts.index, rotation=90)
plt.xlabel("Police Districts")
plt.ylabel("Number of Crimes")
plt.title("Number of Cimes by Police Districts")
plt.show()

#plt.bar(districts)
#crime_district=Crime_df.groupby(["Police District", "Incident Category"])
#print(crime_district["Resolution"].count())
```

```python
#Crime by days
days=Crime_df["Incident Day of Week"].value_counts()
#print(days)
weekday=days["Tuesday"]+days["Monday"]+days["Thursday"]+days["Wednesday"]
weekend=days["Saturday"]+days["Friday"]+days["Sunday"]
#print(days.values)

explode=[0.1,0,0,0,0,0,0]
plt.pie(days, autopct="%1.1f%%", explode=explode,shadow=True, startangle=140, labels=days.index)
plt.show()

weekdays_weekend=[weekday,weekend]
label=["Weekdays", "Weekend"]
explode=[0.1,0]
plt.pie(weekdays_weekend, autopct="%1.1f%%", explode=explode, startangle=120, labels=label)
plt.show()
```

```python
#Crimes by time
Crime_df["Time"]=Crime_df["Incident Time"].replace(":", "", regex=True).astype("float")
#print(Crime_df)
#print(type(Crime_df["Time"].values))
bins=[0.0, 600.0, 1200.0, 1800.0, 2359.0 ]
label=["12am-6am", "6am-12pm", "12pm-6pm", "6pm-12am"]
Crime_df["Time Category"]=pd.cut(Crime_df["Time"],bins,labels=label)
Crime_df.head(25)
crime_by_time=Crime_df["Time Category"].value_counts()
explode=[0.1,0,0,0]

plt.pie(crime_by_time, labels=crime_by_time.index, explode=explode, autopct="%1.1f%%", startangle=140)
plt.show()
```
